```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

## Trazado de matriz de confusi칩n utilizando librer칤a pandas

```python
#Primero se crear치 la matriz de confusi칩n
datos_df = {} #se crea un diccionario
datos_df['Predicted 0'] = [87,17] 

#a침adimos la primera columna a la df
df = pd.DataFrame(data=datos_df)
df
```

```python
#se a침ade la segunda columna al df
datos_df['Predicted 1'] = [16,59]
datos_df
df = pd.DataFrame(data=datos_df)
df
```

```python
#renombramos el nombre de las filas 
df= df.rename(index={0:'Actually 0',1:'Actually 1'})
df
```

```python
#creamos el primer mapa de calor 
heat_map = sns.heatmap(df)
```

```python
#cambiamos color del 'mapa de calor'
heat_map = sns.heatmap(df,cmap='Blues')
```

```python
## se cambian el nombre de las columnas y filas en '0' y '1' para parecerse al gr치fico
df = df.rename(columns={'Predicted 0':'0','Predicted 1':'1'})
df= df.rename(index={'Actually 0':'0','Actually 1':'1'})
```

```python
#trazamos de nuevo el mapa de calor

heat_map = sns.heatmap(df,cmap='Blues', edgecolor = "black",annot=True)
##(annot = True), mostrar치 los n칰meros dentro de cada cuadro 
heat_map = sns.set(font_scale=1) #ajustamos tama침o de la fuente
pass
```

```python
#realizamos cambios 

#trazamos de nuevo el mapa de calor
heat_map = sns.heatmap(df,cmap='Blues', edgecolor = "black",annot=True,linewidths=0.8, linecolor='black')
##(annot = True), mostrar치 los n칰meros dentro de cada cuadro 
heat_map = sns.set(font_scale=1) #ajustamos tama침o de la fuente

plt.xlabel("Predicted label") #cambiamos el valor de 'label' en X
plt.ylabel("True label") #cambiamos el valor de 'label' en y
plt.grid(axis='y')
pass
```

```python
#realizmos cambios finales
#trazamos de nuevo el mapa de calor

#se a침aden 'ticks'
sns.set()
sns.set_style("ticks")

plt.figure(figsize=(10,7))
heat_map = sns.heatmap(df,cmap='Blues', edgecolor = "black",annot=True,linewidths=1, linecolor='k',yticklabels=True)
##(annot = True), mostrar치 los n칰meros dentro de cada cuadro 
heat_map = sns.set(font_scale=1.7) #ajustamos tama침o de la fuente

plt.xlabel("Predicted label",fontsize=15,) #cambiamos el valor de 'label' en X
plt.ylabel("True label",fontsize=15) #cambiamos el valor de 'label' en y
plt.grid(axis='x')
plt.yticks(rotation = 0)
pass
```
